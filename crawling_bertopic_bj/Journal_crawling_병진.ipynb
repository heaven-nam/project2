```python
from requests import get
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
```

```python
url = 'https://scholar.google.com/scholar'
params = {
    'hl':'ko',
    'start':0,
    'as_sdt':'0%2C5',
    'q':'complex analysis',
    'btnG':''
    }
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
```

```python
title_list = []
abstract_list = []
author_list = []
journal_list = []
year_list = []
keyword_list = []
pdf_link_list = []
keyword = ['Keywords' , 'Topics', 'Key Words:', 'Key words', 'IEEE Keywords']
```

```python
pages = list()
for i in range(3):
    pages.append(0+10*i)
for k in range(len(pages)):
    time.sleep(10)
    params['start'] = pages[k]
    resp = get(url, params=params, headers = headers)
    dom = BeautifulSoup(resp.text, 'html.parser')
    divs = dom.find_all('div',attrs={'class':'gs_ri'})
    for div in divs:
        h =  div.find('h3',attrs={'class':'gs_rt'})
        h1 = div.find('div', attrs = {'class':'gs_a'})
        if h.find('span',attrs={'class':'gs_ctc'}) == None : #책 제거
            if h.find('span',attrs={'class':'gs_ctu'}) == None :  #인용 제거
                new_url = h.find('a').attrs['href']
                new_resp = get(new_url, headers = headers)
                new_dom = BeautifulSoup(new_resp.text, 'html.parser')
                
                
                pdf_link_list.append(h.find('a').attrs['href']) #pdf_link(논문의 url임)
                title_list.append(h.find('a').get_text().strip()) #제목
                if h1.find_all('a') == [] :
                    author_list.append(h1.get_text().strip().split('-')[-4:][0].strip())  #저자
                    journal_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[:-1])  #저널
                    year_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[-1].strip())  #년도
                else :
                    author_list.append(h1.get_text().strip().split('-')[-4:][0].strip())  #저자
                    journal_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[:-1])  #저널
                    year_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[-1].strip())  #년도
                for key in new_dom.find_all('ul'):
                    for aaa in keyword :
                        if aaa in key.find_previous().get_text():
                            for iii in key.children:                                   
                                keyword_list.append(iii.get_text()) #keyword 
                                
                for sparse in new_dom.find_all('h2'):
                    if sparse.get_text() == 'Abstract' or sparse.get_text() == 'Summary':
                        abstract_list.append(sparse.find_next().get_text()) #abstract
                

        elif 'HTML' in (h.find('span',attrs={'class':'gs_ctc'})).text : #HTML인 것
                new_url = h.find('a').attrs['href']
                new_resp = get(new_url, headers = headers)
                new_dom = BeautifulSoup(new_resp.text, 'html.parser')
                
                
                pdf_link_list.append(h.find('a').attrs['href']) #pdf_link(논문의 url임)
                title_list.append(h.find('a').get_text().strip()) #제목
                if h1.find_all('a') == [] :
                    author_list.append(h1.get_text().strip().split('-')[-4:][0].strip())  #저자
                    journal_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[:-1])  #저널
                    year_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[-1].strip())  #년도
                else :
                   
                    author_list.append(h1.get_text().strip().split('-')[-4:][0].strip())  #저자
                    journal_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[:-1])  #저널
                    year_list.append(h1.get_text().strip().split('-')[-4:][1].split(',')[-1].strip())  #년도
                for key in new_dom.find_all('h3'):
                    for aaa in keyword :
                        if aaa in key.get_text():
                            for iii in key.find_next().children:
                                keyword_list.append(iii.get_text()) #keyword 
                                
                for sparse in new_dom.find_all('h2'):
                    if sparse.get_text() == 'Abstract' or sparse.get_text() == 'Summary':
                        abstract_list.append(sparse.find_next().get_text()) #abstract


                
      

    
```

```python
from requests import get
from bs4 import BeautifulSoup
import re
import pandas as pd
from urllib.parse import urlencode, parse_qs, urlparse
import time
import random
```

```python
url = 'https://link.springer.com/search/page/2?facet-language=%22En%22&facet-discipline=%22Mathematics%22&query=complex+analysis&facet-content-type=%22Article%22'
parsed = urlparse(url)
qs = parse_qs(parsed.query)
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
```

```python
for k,v in qs.items():
    qs[k] = v[0]
```

```python
qs['query'] = 'real analysis'
```

```python
df = pd.DataFrame(columns=['title','abstract','author','journal','year','pdf_link','keyword'])
```

# real analysis! -> 다음에 할 것, number theory

```python
baseurl = 'https://link.springer.com'

pages = list()
for i in range(200): 
    pages.append(i)
for k in range(len(pages)):
        url = 'https://link.springer.com/search/page/' + str(pages[k]) + '?' + urlencode(qs)
        resp = get(url, headers = headers)
        if resp.status_code != 200:
            print(resp)
            break
        else:
            data = dict()
            dom = BeautifulSoup(resp.text, 'html.parser')
            a = dom.find_all('a',attrs={'class':'title'})
            for h in a:
                link = baseurl + h.attrs['href'] #논문 link 타고 들어가기
                link_resp = get(link, headers = headers)
                link_dom = BeautifulSoup(link_resp.text, 'html.parser')
                if link_resp.status_code != 200:
                    print(link_resp)
                    break
                try :
                    title = link_dom.find('h1', attrs={'class':'c-article-title'}).text
                    data['title'] = [title]
                    abst = link_dom.find('div', attrs={'id':'Abs1-content'})
                    data['abstract'] = [abst.text]
                    author = link_dom.find('ul', attrs={'data-test':'authors-list'})
                    data['author'] = author.text[:-2]
                    journal = link_dom.find('a', attrs={'data-test':'journal-link'}).text
                    data['journal'] = [journal]
                    year = link_dom.find('span', attrs={'data-test':'article-publication-year'}).text
                    data['year'] = [year]
                    data['pdf_link'] = [link]
                    keywords = list()
                    keyul = link_dom.find('ul', attrs={'class':'c-article-subject-list'}).find_all('li', recursive=False)
                    for key in keyul:
                        keywords.append(key.text)
                    data['keyword'] = [','.join(keywords)]
                    data = pd.DataFrame(data)
                    df = pd.concat([df,data])
                except Exception as e:    # 모든 예외의 에러 메시지를 출력할 때는 Exception을 사용
                    print('예외가 발생했습니다.', e)



        sleeping = random.randrange(0,3)
        print('sleep for {}s'.format(sleeping))
        time.sleep(sleeping)
        if k % 20 == 0 and k != 0 : time.sleep(60)
```

```python
from bertopic import BERTopic

model = BERTopic()
topics, probabilities = model.fit_transform(list(df.abstract.values))

print('각 문서의 토픽 번호 리스트 :',len(topics))
print('첫번째 문서의 토픽 번호 :', topics[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english")
model.update_topics(list(df.abstract.values), vectorizer_model=vectorizer_model)
```
