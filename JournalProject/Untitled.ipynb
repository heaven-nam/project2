```python
newcsv = pd.read_csv('1414row.csv')
```

```python
df = pd.DataFrame(newcsv)
```

```python
def removeNewline(x):
    newX = x.split('\n')
    x = ' '.join(newX)
    return x
```

```python
s = df.iloc[0].abstract
```

```python
newdf = df.copy()
newdf.abstract = df.abstract.apply(removeNewline)
```
