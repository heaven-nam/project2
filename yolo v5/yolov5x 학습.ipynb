```python

def labeling(label_list, dd): #jsonÌååÏùºÏïàÏùò bboxÏ†ïÎ≥¥Î•º yolov5Î•º ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌïú txtÌååÏùºÎ°ú Î∞îÍæ∏Îäî Ìï®Ïàò dataÎäî trainÏù∏ÏßÄ valÏù∏ÏßÄÎ•º Ï†ïÌï¥Ï£ºÎäî Ïù∏Ïûê
  for i in range(len(label_list)):
        with open(label_list[i], encoding = 'utf8') as js:
          data = json.load(js)
          data_annot = data['Learning_Data_Info']['Annotations']
          data_type = data_annot[0]['Type']
          img_width = data['Raw_Data_Info']['Resolution'][1]
          img_height = data['Raw_Data_Info']['Resolution'][0]
          if data_type == 'polygon' :
            n = []
            for po in data_annot:
              polygons = po['polygon']

              data_class = po['Class_ID']
              x = []
              y = []
              for t in range(len(polygons)//2):
                x.append(polygons[2*t])
                y.append(polygons[2*t + 1])
              nx = ((min(x) + max(x))/2)/img_width
              ny = ((min(y) + max(y))/2)/img_height
              nw = (max(x) - min(x))/img_width
              nh = (max(y) - min(y))/img_height
              filename = label_list[i].split(".")[0]+".txt"
              n.append([data_class,str(nx),str(ny), str(nw), str(nh)])
          
              
            with open('/content/drive/MyDrive/·Ñá·Öß·Üº·Ñå·Öµ·Ü´/yolov5/important_data/images/' + dd + '/labels/' + filename, 'w') as f:
              for a in n:
                result = ' '.join(s for s in a)
                f.write(result + "\n")
          else: 
            n = []
            for bb in data_annot:
              bbox = bb['bbox']
              data_class = bb['Class_ID']
              bx = ((bbox[0] + bbox[2])/2)/img_width
              by = ((bbox[1] + bbox[3])/2)/img_height
              bw = (bbox[2] - bbox[0])/img_width
              bh = (bbox[3] - bbox[1])/img_height
              filename = label_list[i].split(".")[0]+".txt"
              n.append([data_class,str(bx),str(by), str(bw), str(bh)])
              with open('/content/drive/MyDrive/·Ñá·Öß·Üº·Ñå·Öµ·Ü´/yolov5/important_data/images/' + dd + '/labels/' + filename, 'w') as f:
                for a in n:
                  result = ' '.join(s for s in a)
                  f.write(result + "\n")

        
#polygonÏùÄ bboxÎ°ú Î∞îÍøîÏÑú Ï†ÄÏû•ÌïòÍ≥† bboxÎäî bboxÎì§ÏùÑ Ìïú Î¶¨Ïä§Ìä∏ ÏïàÏóê Îã¥ÏïÑÏÑú Ï†ÄÏû•
```
